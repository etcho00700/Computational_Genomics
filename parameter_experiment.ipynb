```python
def parse_fastq(fh):
    '''
    Parse reads from a FASTQ filehandle. For each read, return a name,
    nucleotide-string, quality-string triple.

    Taken from Professor Langmead.
    '''
    reads = []
    while True:
        first_line = fh.readline()
        if len(first_line) == 0:
            break  # end of file
        name = first_line[1:].rstrip()
        seq = fh.readline().rstrip()
        fh.readline()  # ignore line starting with +
        qual = fh.readline().rstrip()
        reads.append(seq)  # we only care about sequences
    return reads
```

The following code is the main function used to test FPR and runtime over various parameters. The code can be run for sanity check, but all result files are provided in the 'Results' directory in .pkl format. 

You can choose the model (sparse_kbf vs 1_kbf) and the parameter (sizes vs num_hashes vs ks vs skip_lens) by changing the assignments under #CHANGE HERE on line 8. 

```python
fn = path / 'readA.fastq'
fh = open(fn, 'r')
# querying
fastq_reads = parse_fastq(fh)
kmers = set()
setSize = 106432

#CHANGE HERE.
model = '1_kbf' #model = 'sparse_kbf' or model = '1_kbf'
TestParam = 'sizes' # TestParam = 'sizes' or TestParam = 'num_hashes' or TestParam = 'ks' or TestParam = 'skip_lens'


#These parameter values were used. 
sizes = [setSize * i for i in range(1,41,1)]
num_hashes = [2 + i for i in range(0,20,1)]
ks = [2 + i for i in range(30)]
skip_lens = [1,2,3,4,5,6,7,8]
finalData = []

param = sizes
sz = 2000000
nh = 7
ki = 20
sl = 7

if TestParam == 'sizes':
    param = sizes
elif TestParam == 'num_hashes':
    param = num_hashes
elif TestParam == 'ks':
    param = ks
elif TestParam == 'skip_lens':
    param = skip_lens

for p in param:
    # parameters to take from command line later on
    
    size = sz
    num_hash = nh
    k = ki
    skip_len = sl
    extend_len = 2
    
    if TestParam == 'sizes':
        size = p
    elif TestParam == 'num_hashes':
        num_hash = p
    elif TestParam == 'ks':
        k = p
    elif TestParam == 'skip_lens':
        skip_len = p

    false_kmers = list(helper.kmer_generator(10000, k, kmers))
    for read in fastq_reads:
        for i in range(len(read) - k + 1):  # for each k-mer
            kmer = read[i:i + k]
            kmers.add(kmer)

    start = time.time()

    # populate sparse bloom filter
    best_fit_set, edge_set = build_kmer_set.best_fit_kmers(
        fastq_reads, k, skip_len)

    #KBF = sparse_kbf.KBFSparse(size, num_hash, range(
     #   num_hash), k, skip_len, best_fit_set, edge_set)
    KBF = kbf.KmerBloomFilter(size, num_hash, range(num_hash), k, kmers, extend_len)

    true_negative = 0
    false_positive = 0
    for kmer in false_kmers:
        if kmer in KBF:
            false_positive += 1
        else:
            true_negative += 1

    fpr = false_positive / (false_positive + true_negative)

    end = time.time()
    data = [size, num_hash, k, skip_len, extend_len, fpr, end-start]
    finalData.append(data)
    if TestParam == 'sizes':
        print("Size: %d" %(size), "Runtime: ", end - start, "FPR: %.7f" %(fpr))
    elif TestParam == 'num_hashes':
        print("n_hash: %d" %(num_hash), "Runtime: ", end - start, "FPR: %.7f" %(fpr))
    elif TestParam == 'ks':
        print("k: %d" %(k), "FPR: %.7f" %(fpr))
    elif TestParam == 'skip_lens':
        print("skip_len: %d" %(skip_len), "FPR: %.7f" %(fpr))

df = pd.DataFrame(finalData, columns=['bf_size','num_hash','k','skip_len','extend_len','fpr','runtime'])

#homedir = os.path.expanduser("~")
output = model + '_' + TestParam + '.pkl'
df.to_pickle(os.path.join('./Results',output))
```

After the main function is run, the following functions are used to generate graphs for Figure 1. 

```python
plt.rcParams['font.size'] = '14'

sparse_bfsize = pickle.load(open(path / 'Results/sparse_kbf_bfsize.pkl', 'rb'))
kbf_bfsize = pickle.load(open(path / 'Results/1sided_kbf_bfsize.pkl', 'rb'))
sparse_numhash = pickle.load(open(path / 'Results/sparse_kbf_num_hashes.pkl', 'rb'))
kbf_numhash = pickle.load(open(path / 'Results/1sided_kbf_num_hashes.pkl', 'rb'))
sparse_k = pickle.load(open(path / 'Results/sparse_kbf_k.pkl', 'rb'))
kbf_k = pickle.load(open(path / 'Results/1sided_kbf_k.pkl', 'rb'))
sparse_skiplen = pickle.load(open(path / 'Results/sparse_kbf_skip_len.pkl', 'rb'))

def bfsize():

    bf_size1 = sparse_bfsize['bf_size']
    fpr1 = sparse_bfsize['fpr']

    bf_size2 = kbf_bfsize['bf_size']
    fpr2 = kbf_bfsize['fpr']

    plt.plot(bf_size1, fpr1, '-o', label='sparse_kbf')
    plt.plot(bf_size2, fpr2, '-o', label='1_sided_kbf')
    plt.legend()
    plt.xlabel('filter size (x10^6 elements)')
    plt.ylabel('FPR')
    #plt.plot(bf_size, runtime)
    plt.show()

def bfsize_runtime():

    bf_size1 = sparse_bfsize['bf_size']
    runtime1 = sparse_bfsize['runtime']

    bf_size2 = kbf_bfsize['bf_size']
    runtime2 = kbf_bfsize['runtime']

    plt.plot(bf_size1, runtime1, '-o')
    plt.plot(bf_size2, runtime2, '-o')
    plt.xlabel('filter size (x10^6 elements)')
    plt.ylabel('runtime (s)')
    #plt.plot(bf_size, runtime)
    plt.show()

def num_hashes():

    num_hashes1 = sparse_numhash["num_hash"]
    fpr1 = sparse_numhash['fpr']
    num_hashes2 = kbf_numhash["num_hash"]
    fpr2 = kbf_numhash['fpr']

    plt.plot(num_hashes1, fpr1, '-o')
    plt.plot(num_hashes2, fpr2, '-o')
    plt.xlabel("num_hashes")
    plt.ylabel("FPR")
    plt.xticks(range(1,22,2))
    plt.show()

def num_hashes_runtime():
    
    num_hashes1 = sparse_numhash["num_hash"]
    runtime1 = sparse_numhash['runtime']
    num_hashes2 = kbf_numhash["num_hash"]
    runtime2 = kbf_numhash['runtime']

    plt.plot(num_hashes1, runtime1, '-o')
    plt.plot(num_hashes2, runtime2, '-o')
    plt.xlabel("num_hashes")
    plt.ylabel("runtime (s)")
    plt.xticks(range(1,22,2))
    plt.show()

def klen():

    k1= sparse_k["k"]
    fpr1 = sparse_k['fpr']
    k2 = kbf_k["k"]
    fpr2 = kbf_k['fpr']
    
    plt.plot(k1, fpr1, '-o')
    plt.plot(k2, fpr2, '-o')
    plt.xlabel("k length")
    plt.ylabel("FPR")
    plt.show()

def klen_runtime():

    k1= sparse_k["k"]
    runtime1 = sparse_k['runtime']
    k2 = kbf_k["k"]
    runtime2 = kbf_k['runtime']

    plt.plot(k1, runtime1, '-o')
    plt.plot(k2, runtime2, '-o')
    plt.xlabel("k length")
    plt.ylabel("runtime (s)")
    plt.show()

def skiplen():

    skip_len = sparse_skiplen["skip_len"]
    fpr = sparse_skiplen['fpr']

    plt.plot(skip_len, fpr, '-o', label='sparse_kbf')
    plt.xlabel("skip length")
    plt.ylabel("FPR")
    plt.show()

def skiplen_runtime():
    skip_len = sparse_skiplen["skip_len"]
    runtime = sparse_skiplen['runtime']

    plt.plot(skip_len, runtime, '-o', label='sparse_kbf')
    plt.xlabel("skip length")
    plt.ylabel("runtime (s)")
    plt.show()
```
